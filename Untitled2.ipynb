```python
import numpy as np
import cv2
```

```python
with_mask = np.load('with_mask.npy')
without_mask = np.load('without_mask.npy')
```

```python
with_mask.shape
```

```python
without_mask.shape
```

```python
with_mask = with_mask.reshape(200 , 50 * 50 * 3)
```

```python
without_mask = without_mask.reshape(201 , 50 * 50 * 3)
```

```python
with_mask.shape
```

```python
without_mask.shape
```

```python
X = np.r_[with_mask, without_mask]
```

```python
X.shape
```

Training data is completed - X

Label the with mask data to 0

and without mask data to 1

if output is 0 , you wear the mask

if output is 1, you are without mask.

```python
# label the whole data to 0
label = np.zeros(X.shape[0])
```

```python
label
```

```python
# label without mask data with 1

label[200:] = 1.0
```

```python
label
```

Training data is completed

Use Machine learning algorithm to predict the output, we use sklearn pre-built library.

Classification problem because output is in form of 0 and 1.

Here, we use SVM algorithm

```python
# svm - support vector machine

#SVC - Support Vector Classifier


from sklearn.svm import SVC

from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
X
```

```python
label
```

```python
names = {0: 'mask', 1 : 'no_mask'}
```

```python
x_train , x_test, y_train, y_test = train_test_split(X, label, test_size = 0.2)
```

```python
x_train.shape  # 282 image for training
               # remianing image for testing
```

```python
# training dataset 282 rows and 7500 columns
# it is a very big dimension column,it will slow down the algorithm , so we use dimensionality reduction
```

```python
# PCA - Principal component Analysis
from sklearn.decomposition import PCA
```

```python
# transform 7500 into 3 * 3 matrix

pca = PCA(n_components = 3)
```

```python
x_train = pca.fit_transform(x_train)
```

```python
x_test = pca.fit_transform(x_test)
```

```python
x_train
```

```python
x_test.shape
```

```python
x_train.shape     # 282 * 3 matrix
```

```python
y_train.shape
```

Apply SVM algorithm

```python
svm = SVC()

svm.fit(x_train, y_train)
```

```python
# prediction the output
```

```python
y_pred = svm.predict(x_test)
```

```python
y_pred
```

```python
accuracy_score(y_test, y_pred)*100
```

97 percent accuracy is good for model

To use the model

```python
# load the file

haar_data = cv2.CascadeClassifier('data.xml')

# to start the camera
capture = cv2.VideoCapture(0)
# create a data list variaable to store the without mask image
data = []
font = cv2.FONT_HERSHEY_COMPLEX

while True:
    # to read the face
    flag, img = capture.read()
    
    if flag:
        faces = haar_data.detectMultiScale(img)
        for x, y, w, h in faces:
            cv2.rectangle(img,(x,y),(x+w , y+h), (255 , 0 , 255), 4)
            # slicinf the image array
            face = img[y:y+h, x:x+w, :]
            # it detect multiple face so resize it to save all same size faces
            
            face = cv2.resize(face,(50,50))
        
            face = face.reshape(1,-1)
            face = pca.transform(face)
            pred = svm.predict(face)[0]
            
            n = names[int(pred)]
            
            # to write the text on image , cv2 has built in function putText
            cv2.putText(img, n, (x,y), font, 1 , (244 , 255 , 0), 2)
            # 1 represent font size of 16, if 2 then 32 font size
            # 2 represent font-bold
            
            
            print(n)
           
                
        cv2.imshow('result', img)
        if cv2.waitKey(2) == 27: 
                                        # 27 is ASCII code of escape
            break                       # 2 is waiting time in millisecond

capture.release()
cv2.destroyAllWindows()
        
    
```
